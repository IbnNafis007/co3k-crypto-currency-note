```python
# coding=utf-8

import ccxt

exchange = ccxt.kraken()
markets = exchange.load_markets()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.finance as mpf
from datetime import datetime

fig = plt.figure()
ax = plt.subplot()

ohlc = [ [x[0] / 1000] + x[1:] for x in exchange.fetch_ohlcv("BTC/USD", '15m')]
plt.xticks([ohlc[-1][0]], ["aaa"])  # TODO: いい感じの間隔で日付文字列に変換

mpf.candlestick_ohlc(ax, ohlc, width=2, colorup='g', colordown='r')
ax.grid()
```
