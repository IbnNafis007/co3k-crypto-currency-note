```python
# coding=utf-8

import ccxt

exchange = ccxt.kraken()
markets = exchange.load_markets()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.finance as mpf

fig = plt.figure()
ax = plt.subplot()

ohlc = exchange.fetch_ohlcv("BTC/USD", '15m')

mpf.candlestick_ohlc(ax, ohlc, width=2, colorup='g', colordown='r')

ax.grid()
```
